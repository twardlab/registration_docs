### This cell shows an example use case of the slab_alignment script

```python
# Initialize the command string by appending the location of the target script
script_path = os.path.join(os.getcwd(),'slab_alignment.py')
cmd_str = f'python3 {script_path}'

# Define positional and optional arguments (Check docs for more in-depth descriptions for each of these arguments)
dataset = 'TME08-1'
target_files = [os.path.join(os.getcwd(),f'inputs/Camk2a-MORF3-D1Tom_TME08-1_10x_{i:02d}_ch_0_pow_[0.125]_down.npz') for i in range(1,10) ]
tform_file = '/nafs/dtward/andrew_work/test/dragonfly_work/dragonfly_outputs_weighted/TME08-1/dragonfly_output_vis/transformation_outputs.npz'
atlas_names = ['/ifshome/abenneck/docs_public/reg_docs/dendritome_data/atlases/ara_nissl_50.vtk', '/ifshome/abenneck/docs_public/reg_docs/dendritome_data/atlases/average_template_50.vtk', '/ifshome/abenneck/docs_public/reg_docs/dendritome_data/atlases/annotation_50.vtk']
outdir = '/ifshome/abenneck/docs_public/interface_projects/slab_alignment/outputs'
epath = '/ifshome/abenneck/docs_public/reg_docs/emlddmm'
zm = True
debug = True

# Construct the command string using the above arguments
cmd_str += f' {dataset}'

for fname in target_files:
    cmd_str += f' {fname}'

cmd_str += f' -tform_file {tform_file}'

cmd_str += ' -atlas_files'
for fname in atlas_names:
    cmd_str += f' {fname}'

cmd_str += f' -outdir {outdir}'
cmd_str += f' -epath {epath}'

if !zm:
    cmd_str += ' -zm' # Note that the default is False, so passing this flag sets zm to True within the script

if debug:
    cmd_str += ' -debug' # Note that the default is False, so passing this flag sets zm to True within the script

# Print out the example command + then run 
print(cmd_str,'\n')
os.system(cmd_str)
```
